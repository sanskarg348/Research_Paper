```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
```

```python
tweet_list = []
loc = '28.7041, 77.1025, 100km'
for i, tweet in enumerate(sntwitter.TwitterSearchScraper(f' #wastemanagement OR #garbage OR #plastic since:2017-01-01 until:2017-12-31 lang:en geocode:"{loc}"').get_items()):
    if i > 1000:
        break
    tweet_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username,tweet.user.location, tweet.hashtags])
```

```python
df = pd.DataFrame(tweet_list, columns=['Data', 'Id', 'Tweet Text', 'UserName', 'Location', 'Hashtags'])
```
